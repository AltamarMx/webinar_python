```python
from windrose import WindroseAxes
import pandas as pd
import matplotlib.pyplot as plt
```

```python
viento = pd.read_csv('../data/ws_wd.csv',index_col=0,parse_dates=True)
viento.info()
```

```python
viento.plot(subplots=True,figsize=(10,4))
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,6),sharex=True)

ax[0].plot(viento.ws,'k.')
ax[1].plot(viento.wd,'r.')

ax[1].set_ylim(0,360)
ax[1].set_xlabel('Tiempo DD HH:MM')
ax[0].set_ylabel('velocidad ($m/s$)')
ax[1].set_ylabel('Direccion ($^o$)')
```

```python
fig, ax = plt.subplots()
ax.hist(viento.wd, density=True, ec="k")
```

```python
ax = WindroseAxes.from_ax()
ax.bar(viento.wd, viento.ws, normed=True, opening=0.8, edgecolor="white")
ax.set_legend()
```
