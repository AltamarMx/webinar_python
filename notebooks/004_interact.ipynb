```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
import ipywidgets as widgets
```

```python
viento = pd.read_csv("../data/ws_wd.csv", index_col=0, parse_dates=True)
viento = viento[:-1]
```

```python
time=10
resam = viento.resample(f"{time}s").mean()
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(resam.ws, marker="o")
```

```python
def prom_vels(time):
    resam = viento.resample(f"{time}s").mean()
    fig, ax = plt.subplots(figsize=(15,5))
    ax.plot(resam.ws, marker="o")
```

```python
interact(prom_vels, time=[1,5,10,15,30,45,60])
```
