```python
import pandas as pd
import glob as glob
```

```python
f = '../data/03_11_22 [Anemometer 1]-0.txt'
nombres = ['u','v','w','fecha']
tmp = pd.read_csv(f,skiprows=5,usecols=[1,2,3,7],names=nombres)
tmp.fecha = tmp.fecha.str.replace('p. m.','PM', regex=False)
tmp.fecha = tmp.fecha.str.replace('a. m.','AM', regex=False)
tmp.fecha = pd.to_datetime(tmp.fecha, format='%d/%m/%Y %I:%M:%S %p')
tmp.set_index('fecha',inplace=True)
```

```python
def importa(f):
    f = '../data/03_11_22 [Anemometer 1]-0.txt'
    nombres = ['u','v','w','fecha']
    tmp = pd.read_csv(f,skiprows=5,usecols=[1,2,3,7],names=nombres)
    tmp.fecha = tmp.fecha.str.replace('p. m.','PM',regex=True)
    tmp.fecha = tmp.fecha.str.replace('a. m.','AM',regex=True)
    tmp.fecha = pd.to_datetime(tmp.fecha, format='%d/%m/%Y %I:%M:%S %p')
    tmp.set_index('fecha',inplace=True)
    return tmp
```

```python
archivos =  glob.glob('../data/*.txt')
archivos
```

```python
len(archivos)
```

```python
df = pd.concat([importa(archivo) for archivo in archivos]) #tal vez habria que cambiar el list comprehension
```
